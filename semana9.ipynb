### Importando pacotes


```python
import requests as req
import pandas as pd
```

```python
r = req.get('https://www.alura.com.br/api/cursos')
dados=r.json()
#dados3=pd.DataFrame(list(dados))
dados2=pd.json_normalize(data=dados)
dados2
```

```python
unico_slug=dados2.slug.unique()
r = pd.json_normalize(req.get(f'https://cursos.alura.com.br/api/curso-{unico_slug[0]}').json(),max_level=0)

for i in range(1,len(unico_slug)):
    r2 = pd.json_normalize(req.get(f'https://cursos.alura.com.br/api/curso-{unico_slug[i]}').json(),max_level=0)
    r=pd.concat([r,r2])
```

```python
pd.set_option("display.max_columns",50)
r
```

```python
lista=['ementa',
'chamadas',
'publico_alvo',
'video_1a_aula',
'instrutores',
'musicas',
'depoimentos',
'transcriptions',
'highlightedExtraField',
'curso_substituto',
'meta_title',
'video_1a_aula',
'parceiros',
'formacoes']
```

```python
r.drop(lista,axis=1,inplace=True)
r
```

```python
pd.reset_option("display.max_columns")
```

```python
r['alunos_ativos_total']=pd.json_normalize(r.alunos_ativos)['total']
r['subcategoria_nome']=pd.json_normalize(r.subcategoria)['nome']
r['categoria_nome']=pd.json_normalize(r.categoria)['nome']
r
```
